#### Objective Function:
Minimize soft constraint penalties while satisfying all hard constraints.

#### Hard Constraints:
---

#### A. Every shift must have exactly one pediatrician:

   For every \( j \), the sum of \( X_{i,j} \) over all \( i \) must equal 1:
   $$
   \sum_{i} X_{i,j} = 1, \quad \forall j
   $$

#### B. Pediatricians’ Predetermined Number of Shifts
 The predetermined number of shifts for pediatricians is listed in the Excel file ‘BCWH_Inputs’ under cells E62:S62 as ‘Total shifts required to work for this cycle.’ Let R_{i,j} denote the availability of pediatrician j for shift i, where 1R_{i,j}=1 indicates availability and R_{i,j}=0 indicates unavailability. The ‘Total shifts required to work for this cycle’ is calculated as the sum of R_{i,j}over all i. Additionally, cells E4:S4 represent pediatrician j, cells D5:D60 list shifts i, and cells E5:S60 detail pediatrician j’s availability for each shift i.

- Let \( R_{i,j} \) denote the availability of pediatrician \( j \) for shift \( i \), where:
  - \( R_{i,j} = 1 \): Indicates availability.
  - \( R_{i,j} = 0 \): Indicates unavailability.
  
- The **‘Total shifts required to work for this cycle’** for each pediatrician \( j \) is calculated as:
  $$
  T_j = \sum_{i} R_{i,j}, \quad \forall j
  $$

- \( i \): Indexes the shifts (from cells \( D5:D60 \)).
- \( j \): Indexes the pediatricians (from cells \( E4:S4 \)).
- \( R_{i,j} \): Pediatrician \( j \)’s availability for shift \( i \) (from cells \( E5:S60 \)).
- \( T_j \): The total number of shifts required for pediatrician \( j \) (from \( E62:S62 \)).

---

#### C. No Back-to-Back Shifts
To ensure no pediatrician is assigned to consecutive shifts:
- Let \( X_{i,j} \) represent whether pediatrician \( j \) is assigned to shift \( i \), where:
  - \( X_{i,j} = 1 \): Assigned.
  - \( X_{i,j} = 0 \): Not assigned.
  
- The constraint is expressed as:
  $$
  X_{i,j} + X_{i+1,j} \leq 1, \quad \forall j, \, \forall i
  $$

---

#### D. No Consecutive Night Shifts
To prevent assignments to two consecutive even-numbered shifts:
- Let \( X_{i,j} \) represent whether pediatrician \( j \) is assigned to shift \( i \), where:
  - \( X_{i,j} = 1 \): Assigned.
  - \( X_{i,j} = 0 \): Not assigned.
  
- The constraint is expressed as:
  $$
  X_{i,j} + X_{i+2,j} \leq 1, \quad \forall j, \, \forall i \text{ where } i \text{ is even.}
  $$

For example:
- \( X_{2,j} + X_{4,j} \leq 1 \)
- \( X_{4,j} + X_{6,j} \leq 1 \), and so on.


#### E. No More Than Two Shifts in a Weekend
To ensure no pediatrician works more than two shifts in a weekend:

- Let \( X_{i,j} \) represent whether pediatrician \( j \) is assigned to shift \( i \).
The constraint for week 1 is expressed as:

  $$
  \sum_{i=10}^{14} X_{i,j} \leq 2, \quad \forall j
  $$
 
This condition repeats for subsequent weekends:

- For Week 2 (Shifts 24–28):
  $$
  \sum_{i=24}^{28} X_{i,j} \leq 2, \quad \forall j
  $$

- For Week 3 (Shifts 38–42):
  $$
  \sum_{i=38}^{42} X_{i,j} \leq 2, \quad \forall j
  $$

- For Week 4 (Shifts 52–56):
  $$
  \sum_{i=52}^{56} X_{i,j} \leq 2, \quad \forall j
  $$

---

#### F. Pediatricians Should Not Work Shifts They Requested Off
To ensure pediatricians are not assigned shifts they have requested off:
Shift requests are in Excel file ‘BCWH_Inputs’. cell D5:D60 label shift ‘i’, E5:S60 contains pediatrician ‘j’s availability for shift ‘i’.

- Let \( R_{i,j} \) represent pediatrician \( j \)’s availability for shift \( i \), where:
  - \( R_{i,j} = 1 \): Available.
  - \( R_{i,j} = 0 \): Not available (requested off).
- The constraint is:
  $$
  X_{i,j} \leq R_{i,j}, \quad \forall i, \forall j
  $$

---

#### G. POW (Person on Watch) Requirement
To ensure proper assignment of POW shifts:

- Let \( Z_{w,j} \) be a binary variable indicating whether pediatrician \( j \) is assigned as POW during week \( w \).
- For each week, the constraint is:
  $$
  X_{3,j} + X_{7,j} + X_{12,j} = 3 \cdot Z_{w,j}, \quad \forall j
  $$
  Similarly:
  $$
  X_{17,j} + X_{21,j} + X_{26,j} = 3 \cdot Z_{w,j}, \quad \forall j
  $$

---

#### I. POW No More Than Once in a Four-Week Cycle
To limit POW assignments to once per four-week cycle:

- The constraint is:
  $$
  Z_{1,j} + Z_{2,j} + Z_{3,j} + Z_{4,j} \leq 1, \quad \forall j
  $$

---

#### J. Remaining POW Requirements
To meet the total remaining POW requirements for the year:

- Let \( P_j \) represent the remaining number of POW shifts pediatrician \( j \) must complete.
- The constraint is:
  $$
  Z_{1,j} + Z_{2,j} + Z_{3,j} + Z_{4,j} \leq P_j, \quad \forall j
  $$

### Soft Constraints

#### A. Avoid Consecutive Weekend Shifts
To discourage assigning pediatricians to consecutive weekend shifts:

- Let \( Y_{w,j} \) indicate if pediatrician \( j \) worked on the weekend of week \( w \).
- Let \( U_j \) and \( O_j \) represent penalties for under- and over-assignments.
- For each weekend:
  $$
  \sum_{i=10}^{14} X_{i,j} + U_j + O_j \leq 5 \cdot Y_{1,j}, \quad \forall j
  $$

Similar constraints apply for other weekends:

- **Week 2 (Shifts 24–28):**
  $$
  \sum_{i=24}^{28} X_{i,j} + U_j + O_j \leq 5 \cdot Y_{2,j}, \quad \forall j
  $$

- **Week 3 (Shifts 38–42):**
  $$
  \sum_{i=38}^{42} X_{i,j} + U_j + O_j \leq 5 \cdot Y_{3,j}, \quad \forall j
  $$

- **Week 4 (Shifts 52–56):**
  $$
  \sum_{i=52}^{56} X_{i,j} + U_j + O_j \leq 5 \cdot Y_{4,j}, \quad \forall j
  $$

- Additionally:
  $$
  Y_{1,j} + Y_{2,j} + U_j + O_j \leq 1, \quad \forall j
  $$
  And for subsequent weekends:
  $$
  Y_{2,j} + Y_{3,j} + U_j + O_j \leq 1, \quad Y_{3,j} + Y_{4,j} + U_j + O_j \leq 1, \quad \forall j
  $$

---

#### B. Avoid Night Shifts Two Days Apart
To avoid assigning night shifts that are two days apart:

- The constraint is:
  $$
  X_{2,j} + X_{4,j} + X_{6,j} \leq 1, \quad \forall j
  $$
  Similarly:
  $$
  X_{4,j} + X_{6,j} + X_{8,j} + U_j + O_j \leq 1, \quad \forall j
  $$

---

#### C. Limit Evening Shifts Per Week
To limit the number of evening shifts per week:

- For each week:
  $$
  \sum_{i \in \text{even indices}} X_{i,j} + U_j + O_j \leq 2, \quad \forall j
  $$

---

#### D. Equity in Day vs. Night Shifts
To maintain balance between day and night shifts for each pediatrician:

- Even-indexed shifts are night shifts, and odd-indexed shifts are day shifts:

  $$
  \sum_{i \in \text{even indices}} X_{i,j} + O_j + U_j = \sum_{i \in \text{odd indices}} X_{i,j}, \quad \forall j
  $$

```python
from gurobipy import Model, GRB, quicksum
import pandas as pd
import os

# Load the Excel file
file_name = "BCWH_Inputs.xlsx"  # Excel file name
file_path = os.path.join(os.getcwd(), file_name)  # Build full file path

# Read the data directly, targeting the range E4:S60
data = pd.read_excel(file_path, sheet_name="Sheet1", usecols="E:S", skiprows=3, nrows=57)  # Skipping rows above E4

# Convert to numeric and handle any non-numeric values
availability_data = data.apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)
availability = availability_data.values  # Convert to a NumPy-like 2D array

# Debugging: Verify the availability matrix
shifts = list(range(1, 57))  # Shift numbers 1 to 56
pediatricians = list(range(1, 16))  # Pediatricians 1 to 15

# Hard-code "Total shifts required to work for this cycle"
shift_requirements = [6,2,4,4,3,3,4,3,6,3,4,3,4,4,3]  # Example values, replace with actual requirements

shifts = list(range(1, 57))  # Shift numbers 1 to 56
pediatricians = list(range(1, 16))  # Pediatricians 1 to 15

# Define constants
POW_shifts = {1: [3, 7, 12], 2: [17, 21, 26], 3: [31, 35, 40], 4: [45, 49, 54]}  # POW required shifts
weekends = [(10, 14), (24, 28), (38, 42), (52, 56)]  # Weekend ranges for shifts
remaining_pow = [1, 2, 0, 1, 0, 3, 1, 1, 1, 2, 1, 3, 0, 0, 1]  # Update with actual remaining POW data

# Create model
model = Model("Pediatrician Scheduling")

# Decision variables
X = model.addVars(shifts, pediatricians, vtype=GRB.BINARY, name="X")  # Assign pediatricians to shifts
Z = model.addVars(range(1, 5), pediatricians, vtype=GRB.BINARY, name="Z")  # POW assignments

# Soft constraint penalty variables
U = model.addVars(pediatricians, vtype=GRB.CONTINUOUS, name="UnderPenalties")  # Under penalties
O = model.addVars(pediatricians, vtype=GRB.CONTINUOUS, name="OverPenalties")  # Over penalties

U_b = model.addVars(pediatricians, vtype=GRB.CONTINUOUS, name="UnderPenaltyNightSeparation")
O_b = model.addVars(pediatricians, vtype=GRB.CONTINUOUS, name="OverPenaltyNightSeparation")

U_c = model.addVars(pediatricians, vtype=GRB.CONTINUOUS, name="UnderPenaltyEveningLimit")
O_c = model.addVars(pediatricians, vtype=GRB.CONTINUOUS, name="OverPenaltyEveningLimit")

U_d = model.addVars(pediatricians, vtype=GRB.CONTINUOUS, name="UnderPenaltyDayNightEquity")
O_d = model.addVars(pediatricians, vtype=GRB.CONTINUOUS, name="OverPenaltyDayNightEquity")

# Objective function with penalties
penalty_weights = {
    "consecutive_weekends": 10,
    "night_shift_separation": 8,
    "evening_shift_limit": 5,
    "day_night_equity": 3
}

model.setObjective(
    quicksum(U[j] * penalty_weights["consecutive_weekends"] for j in pediatricians) +
    quicksum(O[j] * penalty_weights["consecutive_weekends"] for j in pediatricians) +
    quicksum(U_b[j] * penalty_weights["night_shift_separation"] for j in pediatricians) +
    quicksum(O_b[j] * penalty_weights["night_shift_separation"] for j in pediatricians) +
    quicksum(U_c[j] * penalty_weights["evening_shift_limit"] for j in pediatricians) +
    quicksum(O_c[j] * penalty_weights["evening_shift_limit"] for j in pediatricians) +
    quicksum(U_d[j] * penalty_weights["day_night_equity"] for j in pediatricians) +
    quicksum(O_d[j] * penalty_weights["day_night_equity"] for j in pediatricians),
    GRB.MINIMIZE
)

# Hard constraints

# A. Every shift must have exactly one pediatrician
for i in shifts:
    model.addConstr(quicksum(X[i, j] for j in pediatricians) == 1, name=f"Shift_{i}_OnePediatrician")

# B. Pediatricians' predetermined number of shifts
for j in pediatricians:
    model.addConstr(quicksum(X[i, j] for i in shifts) == shift_requirements[j - 1], name=f"Pediatrician_{j}_ShiftCount")

# C. No back-to-back shifts
for j in pediatricians:
    for i in range(1, 56):
        model.addConstr(X[i, j] + X[i + 1, j] <= 1, name=f"BackToBack_{i}_{j}")

# D. No consecutive night shifts
for j in pediatricians:
    for i in range(2, 56, 2):  # Even-indexed shifts
        if i + 2 <= 56:
            model.addConstr(X[i, j] + X[i + 2, j] <= 1, name=f"ConsecutiveNight_{i}_{j}")

# E. No more than two shifts in a weekend
for j in pediatricians:
    for start, end in weekends:
        model.addConstr(quicksum(X[i, j] for i in range(start, end + 1)) <= 2, name=f"Weekend_{start}_{j}")

# F. Pediatricians should not work on shifts they requested off
for i, shift_availability in enumerate(availability, start=1):  # Ensure shift indexing starts at 1
    for j, available in enumerate(shift_availability, start=1):  # Ensure pediatrician indexing starts at 1
        if not available:  # If unavailable
            model.addConstr(X[i, j] == 0, name=f"Unavailable_Shift_{i}_Pediatrician_{j}")

# G. POW requirements
for w in range(1, 5):  # For each week
    for j in pediatricians:  # For each pediatrician
        model.addConstr(quicksum(X[s, j] for s in POW_shifts[w]) == 3 * Z[w, j], name=f"POW_Shifts_{w}_{j}")

# I. No one serves as POW more than once in a cycle
for j in pediatricians:
    model.addConstr(quicksum(Z[w, j] for w in range(1, 5)) <= 1, name=f"POW_Limit_{j}")

# J. POW remaining assignment constraint
for j in pediatricians:
    model.addConstr(quicksum(Z[w, j] for w in range(1, 5)) <= remaining_pow[j - 1], name=f"POW_Remaining_{j}")

# Soft constraints

# A. Not to work on consecutive weekends
for j in pediatricians:
    for w in range(1, 4):  # Weeks 1 through 3
        model.addConstr(
            quicksum(X[s, j] for s in range(weekends[w - 1][0], weekends[w - 1][1] + 1)) +
            quicksum(X[s, j] for s in range(weekends[w][0], weekends[w][1] + 1)) <= 1 + U[j] + O[j],
            name=f"ConsecutiveWeekends_{w}_{j}"
        )

# B. Not to work night shifts two days apart
for j in pediatricians:
    for i in range(2, 56, 2):  # Night shifts (even-indexed)
        if i + 2 <= 56:  # Ensure no index out of bounds
            model.addConstr(
                X[i, j] + X[i + 2, j] <= 1 + U_b[j] + O_b[j], name=f"NightShiftSeparation_{i}_{j}"
            )

# C. No more than two evening shifts in a given week
for j in pediatricians:
    for w in range(1, 5):  # Weeks 1 to 4
        week_start = (w - 1) * 14 + 2  # Start of night shifts for week w
        week_night_shifts = range(week_start, week_start + 14, 2)  # All night shifts in the week
        model.addConstr(
            quicksum(X[i, j] for i in week_night_shifts) <= 2 + U_c[j] + O_c[j], name=f"EveningShiftLimit_{w}_{j}"
        )

# D. Equity in day versus night shifts
for j in pediatricians:
    day_shifts = range(1, 57, 2)  # Odd-indexed shifts are day shifts
    night_shifts = range(2, 57, 2)  # Even-indexed shifts are night shifts
    model.addConstr(
        quicksum(X[i, j] for i in night_shifts) - quicksum(X[i, j] for i in day_shifts) <= U_d[j] + O_d[j],
        name=f"DayNightEquity_Over_{j}"
    )
    model.addConstr(
        quicksum(X[i, j] for i in day_shifts) - quicksum(X[i, j] for i in night_shifts) <= U_d[j] + O_d[j],
        name=f"DayNightEquity_Under_{j}"
    )

# Solve the model
model.optimize()

# Output the results
if model.status == GRB.OPTIMAL:
    schedule = {(i, j): X[i, j].x for i in shifts for j in pediatricians if X[i, j].x > 0.5}
    pow_schedule = {(w, j): Z[w, j].x for w in range(1, 5) for j in pediatricians if Z[w, j].x > 0.5}
    
    # Convert to DataFrame for better readability (optional)
    schedule_df = pd.DataFrame(schedule.keys(), columns=["Shift", "Pediatrician"])
    pow_df = pd.DataFrame(pow_schedule.keys(), columns=["Week", "Pediatrician"])
    print("Shift Assignments:")
    print(schedule_df)
    print("POW Assignments:")
    print(pow_df)
else:
    print("No optimal solution found.")
```
